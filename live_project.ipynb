## Task 1

```python
import pandas as pd
import numpy as np
df = pd.read_csv("bangalore_tech_salaries (1).csv")
df.head()
print(df.shape)
print(df.columns)
df.info()
print(df.dtypes)
print(df.isnull().sum())
print(df.duplicated().sum())
df.describe()
```

## Task 2: Data Cleaning

```python
# Convert column names into snake_case format
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
)
print(df.columns)
# Display all unique role names
print(df["role"].unique())
print(df["company_type"].unique())
print(df["education_tier"].unique())
print(df["work_mode"].unique())

# Create a cleaned role column
role_mapping = {
    "Backend Dev": "SDE Backend",
    "Backend Developer": "SDE Backend",
    "Backend Engineer": "SDE Backend",
    "BE": "SDE Backend",
    "SDE-Backend": "SDE Backend",

    "Frontend Dev": "SDE Frontend",
    "Frontend Developer": "SDE Frontend",
    "Frontend Engineer": "SDE Frontend",
    "FE": "SDE Frontend",
    "SDE-Frontend": "SDE Frontend",

    "FullStack": "SDE Full-Stack",
    "Fullstack Dev": "SDE Full-Stack",
    "Full Stack Engineer": "SDE Full-Stack",
    "SDE FS": "SDE Full-Stack",

    "DS": "Data Scientist",
    "Data Science Engineer": "Data Scientist",
    "ML Engineer": "Data Scientist",

    "DA": "Data Analyst",
    "BI Analyst": "Data Analyst",
    "Analytics Engineer": "Data Analyst",

    "BA": "Business Analyst",
    "Business Systems Analyst": "Business Analyst",

    "PM": "Product Manager",
    "Sr PM": "Product Manager",
    "Product Lead": "Product Manager",

    "UI Designer": "UI/UX",
    "UX Designer": "UI/UX",
    "Designer": "UI/UX",
    "Product Designer": "UI/UX",

    "DevOps": "DevOps Engineer",

    "Infra Engineer": "Site Reliability Engineer",
    "SRE": "Site Reliability Engineer"
}

df["role_clean"] = df["role"].replace(role_mapping)
print(df["role_clean"].value_counts())

# Standardize company type names
company_mapping = {
    "mnc": "MNC",
    "unicorn": "Unicorn",
    "UNICORN": "Unicorn",
    "mid-size": "Mid-size",
    "MID-SIZE": "Mid-size",
    "early-stage": "Early-stage",
    "EARLY-STAGE": "Early-stage"
}

df["company_type_clean"] = df["company_type"].replace(company_mapping)
print(df["company_type_clean"].value_counts())

# Standardize education tier values
education_mapping = {
    "Tier-1": "Tier 1",
    "T1": "Tier 1",
    "1": "Tier 1",

    "Tier-2": "Tier 2",
    "T2": "Tier 2",
    "2": "Tier 2",

    "Tier-3": "Tier 3",
    "T3": "Tier 3",
    "3": "Tier 3"
}

df["education_tier_clean"] = df["education_tier"].replace(education_mapping)
print(df["education_tier_clean"].value_counts())

# Standardize work mode values
work_mode_mapping = {
    "WFH": "Remote",
    "Work from Office": "Onsite",
    "Hybrid (3 days)": "Hybrid"
}

df["work_mode_clean"] = df["work_mode"].replace(work_mode_mapping)
print(df["work_mode_clean"].value_counts())

# Check different salary formats present in Current_CTC
print(df["current_ctc"].head(20))
print(df["previous_ctc"].head(20))

# Convert different salary formats into LPA (float)
def parse_ctc(value):

    # Handle missing values
    if pd.isna(value):
        return np.nan

    # Convert everything to string
    value = str(value).strip()

    # Remove currency symbol and LPA text
    value = value.replace("₹", "")
    value = value.replace("LPA", "")
    value = value.replace("lpa", "")
    value = value.strip()

    # Case 1: Salary stored as 15,50,000
    if "," in value:
        salary = float(value.replace(",", ""))

        # Convert rupees into LPA
        if salary > 100000:
            return salary / 100000

    # Case 2: Already stored as 15.5
    return float(value)
df["current_ctc_lpa"] = df["current_ctc"].apply(parse_ctc)

df["previous_ctc_lpa"] = df["previous_ctc"].apply(parse_ctc)

df[["current_ctc", "current_ctc_lpa"]].head(15)

print(df[["current_ctc_lpa", "previous_ctc_lpa"]].isnull().sum())

df = df.drop_duplicates()

print("Dataset shape after removing duplicates:", df.shape)

print(df.dtypes)
print(df["role_clean"].value_counts())
print(df["company_type_clean"].value_counts())
print(df["education_tier_clean"].value_counts())
df[["current_ctc", "current_ctc_lpa"]].head(10)
```

## Task 3.1: CTC Distribution by Role

```python
# Calculate salary statistics for each cleaned role

role_ctc = (
    df.groupby("role_clean")["current_ctc_lpa"]
      .agg(
          Median_CTC="median",
          Mean_CTC="mean",
          Min_CTC="min",
          Max_CTC="max"
      )
      .sort_values(by="Median_CTC", ascending=False)
)

print(role_ctc)
Median_CTC="median"
Mean_CTC="mean"
Min_CTC="min"
Max_CTC="max"
```

### Insight (Q3.1)

- Product Manager has the highest median salary at **31.30 LPA**, making it the highest-paying role in this dataset.
- Data Analyst has the lowest median salary at **16.90 LPA**, which is significantly lower than most engineering roles.
- Roles such as Product Manager and SDE Full-Stack show a noticeable gap between the mean and median CTC, suggesting the presence of high-salary outliers.

##Task 3.2 – Experience Curve for SDE Backend

```python
# Filter only SDE Backend employees
backend_df = df[df["role_clean"] == "SDE Backend"].copy()

# Create experience groups
bins = [-1, 1, 3, 5, 100]
labels = ["0-1 Years", "2-3 Years", "4-5 Years", "6+ Years"]

backend_df["exp_band"] = pd.cut(
    backend_df["years_exp"],
    bins=bins,
    labels=labels
)

# Calculate median CTC for each experience band
experience_ctc = (
    backend_df
    .groupby("exp_band", observed=False)["current_ctc_lpa"]
    .median()
    .reset_index()
)

# Calculate percentage growth
experience_ctc["Growth_%"] = (
    experience_ctc["current_ctc_lpa"]
    .pct_change() * 100
)

print(experience_ctc)
```

### Insight (Q3.2)

- The median salary for SDE Backend professionals increases from **11.65 LPA** (0–1 years) to **40.40 LPA** (6+ years).
- The highest salary growth (**71.67%**) occurs between **0–1 years** and **2–3 years**, indicating rapid early-career progression.
- Professionals who build strong technical skills and gain experience during their first three years can significantly improve their earning potential.

##Task 3.3 – Skill Premium for SDEs

```python
# Filter only SDE roles
sde_df = df[df["role_clean"].isin([
    "SDE Backend",
    "SDE Frontend",
    "SDE Full-Stack"
])].copy()

# Skills to compare
skills = ["AWS", "ML", "System Design", "Kubernetes"]

# Store results
results = []

# Compare salary for employees with and without each skill
for skill in skills:

    with_skill = sde_df[
        sde_df["skills"].str.contains(skill, case=False, na=False)
    ]

    without_skill = sde_df[
        ~sde_df["skills"].str.contains(skill, case=False, na=False)
    ]

    median_with = with_skill["current_ctc_lpa"].median()
    median_without = without_skill["current_ctc_lpa"].median()

    premium = ((median_with - median_without) / median_without) * 100

    results.append([
        skill,
        round(median_with, 2),
        round(median_without, 2),
        round(premium, 2)
    ])

# Create DataFrame
skill_premium = pd.DataFrame(
    results,
    columns=[
        "Skill",
        "Median With Skill",
        "Median Without Skill",
        "Premium (%)"
    ]
)

# Sort by premium
skill_premium = skill_premium.sort_values(
    by="Premium (%)",
    ascending=False
)

print(skill_premium)
```

### Insight (Q3.3)

- Among the selected skills, **System Design** provides the highest salary premium, with professionals earning **20.48%** more than those without this skill.
- **Machine Learning (ML)** also offers a noticeable salary advantage of **12.44%**, while **Kubernetes** provides a moderate premium of **7.91%**.
- Interestingly, professionals with **AWS** have a **4.34% lower** median salary than those without AWS in this dataset, indicating that AWS alone may not be a strong salary differentiator and is often expected as a common skill.

## Task 3.4 – Company Type Premium

```python
# Filter only SDE Backend employees
backend_df = df[df["role_clean"] == "SDE Backend"].copy()

# Calculate median salary by company type
company_ctc = (
    backend_df
    .groupby("company_type_clean")["current_ctc_lpa"]
    .median()
    .reset_index()
)

print(company_ctc)

# Get median salaries
unicorn_salary = company_ctc.loc[
    company_ctc["company_type_clean"] == "Unicorn",
    "current_ctc_lpa"
].values[0]

mnc_salary = company_ctc.loc[
    company_ctc["company_type_clean"] == "MNC",
    "current_ctc_lpa"
].values[0]

# Calculate percentage premium
premium = ((unicorn_salary - mnc_salary) / mnc_salary) * 100

print(f"Unicorn Premium over MNC: {premium:.2f}%")
unicorn_salary = 26
mnc_salary = 20
```

### Insight (Q3.4)

- Among SDE Backend professionals, **Unicorn companies offer the highest median salary of 27.35 LPA**, while **Early-stage companies offer the lowest median salary of 18.60 LPA**.
- The median salary in **Unicorn companies is 33.74% higher than in MNCs**, indicating a significant company-type premium for the same role.
- Candidates seeking higher compensation may benefit from targeting Unicorn companies, while also considering factors such as job stability, learning opportunities, and work-life balance.

##Task 3.5 – Top 10 Most Underpaid Professionals

```python
# Create experience bands for the complete dataset

bins = [-1, 1, 3, 5, 100]

labels = ["0-1 Years", "2-3 Years", "4-5 Years", "6+ Years"]

df["exp_band"] = pd.cut(
    df["years_exp"],
    bins=bins,
    labels=labels
)
# Count employees in each group

group_size = (
    df.groupby(
        ["role_clean", "company_type_clean", "exp_band"],
        observed=False
    )["employee_id"]
    .transform("count")
)

df["group_size"] = group_size
# Calculate median salary within each group

df["group_median"] = (
    df.groupby(
        ["role_clean", "company_type_clean", "exp_band"],
        observed=False
    )["current_ctc_lpa"]
    .transform("median")
)
# Difference between employee salary and group median

df["gap"] = (
    df["current_ctc_lpa"] -
    df["group_median"]
)
# Keep groups with at least 10 employees

underpaid_df = df[
    df["group_size"] >= 10
]
# Sort employees by salary gap

underpaid_df = underpaid_df.sort_values(
    by="gap"
)
# Display top 10 most underpaid professionals

top10 = underpaid_df[
    [
        "employee_id",
        "role_clean",
        "company_type_clean",
        "years_exp",
        "current_ctc_lpa",
        "group_median",
        "gap"
    ]
].head(10)

print(top10)
```

### Insight (Q3.5)

- The most underpaid employee in the dataset is **BLR0615 (Data Scientist)**, earning **28.10 LPA**, which is **12.20 LPA below** the median salary of professionals with the same role, company type, and experience level.
- Several employees have salaries significantly below their respective group medians, indicating noticeable compensation differences even among professionals with similar roles, experience, and company types.
- This analysis helps identify employees who may be underpaid relative to their peers and can support salary benchmarking and compensation review decisions.
